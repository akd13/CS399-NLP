```python
charts = ['statista_images', 'pew_images']
dataset = '/dataset'
multicolumn = '/multiColumn' #for multiColumn data
images = '/imgs'
descriptions = '/captions/' #human written descriptions are titled captions
#alt text has been extracted as title for Pew. Paper unclear about exactly is title. We use it as captions
captions = '/titles/' 
context = '/data' #we use underlying data tables as context
```

# Statista

```python
df = pd.read_csv(charts[0]+dataset+'/metadata.csv')
```

```python
for key, value in df.iloc[0].items():
    print(key, ":",value)
```

```python
df.head(5)
```

```python
statista_json = {'images':[]}
```

```python
for index, raw_entry in df.iterrows():
    json_entry = {}
    entry = raw_entry.to_dict()
    json_entry['article_id'] = entry['Region']
    json_entry['filename'] = entry['imgPath'].split('/')[-1]
    json_entry['orig_filename'] = entry['URL']
    json_entry['description'] = { 'raw': entry['first_caption'], 'tokens': word_tokenize(entry['first_caption'])}
    raw_caption = entry['title'].replace('\n','').strip()
    json_entry['caption'] = { 'raw': raw_caption, 'tokens': word_tokenize(raw_caption)}
    raw_context = entry['caption'].replace(entry['first_caption'],"")
    json_entry['context'] = { 'raw': raw_context, 'tokens': word_tokenize(raw_context)}
    statista_json['images'].append(json_entry)
    #TODO: Add underlying data
```

```python
df_multi = pd.read_csv(charts[0]+dataset+'/multiColumn/metadata.csv')
```

```python
for index, raw_entry in df_multi.iterrows():
    json_entry = {}
    entry = raw_entry.to_dict()
    json_entry['article_id'] = entry['Region']
    json_entry['filename'] = 'multiColumn'+entry['imgPath'].split('/')[-1]
    json_entry['orig_filename'] = entry['URL']
    json_entry['description'] = { 'raw': entry['first_caption'], 'tokens': word_tokenize(entry['first_caption'])}
    raw_caption = entry['title'].replace('\n','').strip()
    json_entry['caption'] = { 'raw': raw_caption, 'tokens': word_tokenize(raw_caption)}
    raw_context = entry['caption'].replace(entry['first_caption'],"")
    json_entry['context'] = { 'raw': raw_context, 'tokens': word_tokenize(raw_context)}
    statista_json['images'].append(json_entry)
    #TODO: Add underlying data
```

```python
with open("statista_images/statista.json", "w") as outfile:
    outfile.write(json.dumps(statista_json))
```

# Pew

```python
df_pew = pd.read_csv(charts[1]+dataset+'/metadata.csv')
```

```python
df_pew.head(5)
```

```python
pew_json = {'images':[]}
```

```python
for index, raw_entry in df_pew.iterrows():
    json_entry = {}
    entry = raw_entry.to_dict()
    json_entry['article_id'] = entry['topic']
    json_entry['filename'] =entry['imgPath'].split('/')[-1]
    json_entry['orig_filename'] = entry['URL']
    json_entry['description'] = { 'raw': entry['caption'], 'tokens': word_tokenize(entry['caption'])}
    raw_caption = entry['title'].replace('\n','').strip()
    json_entry['caption'] = { 'raw': raw_caption, 'tokens': word_tokenize(raw_caption)}
    raw_context = str(open('pew_images/dataset/'+entry['dataPath'],"r").readlines())
    json_entry['context'] = { 'raw': raw_context, 'tokens': word_tokenize(raw_context)}
    pew_json['images'].append(json_entry)
```

```python
len(pew_json['images'])
```

```python
df_pew_multicolumn = pd.read_csv(charts[1]+dataset+'/multiColumn/metadata.csv')
```

```python
df_pew_multicolumn.head(5)
```

```python
for index, raw_entry in df_pew_multicolumn.iterrows():
    json_entry = {}
    entry = raw_entry.to_dict()
    json_entry['article_id'] = entry['topic']
    json_entry['filename'] ='mutliColumn'+entry['imgPath'].split('/')[-1]
    json_entry['orig_filename'] = entry['URL']
    json_entry['description'] = { 'raw': entry['caption'], 'tokens': word_tokenize(entry['caption'])}
    raw_caption = entry['title'].replace('\n','').strip()
    json_entry['caption'] = { 'raw': raw_caption, 'tokens': word_tokenize(raw_caption)}
    raw_context = str(open('pew_images/dataset/'+entry['dataPath'],"r").readlines()[0])
    json_entry['context'] = { 'raw': raw_context, 'tokens': word_tokenize(raw_context)}
    pew_json['images'].append(json_entry)
```

```python
len(pew_json['images'])
```

```python
with open("pew_images/pew.json", "w") as outfile:
    outfile.write(json.dumps(pew_json))
```
